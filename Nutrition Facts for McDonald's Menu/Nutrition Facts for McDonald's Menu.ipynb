```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'],bins = 30, kde = True)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.axvline(x=500,color='y',label='500 Calories',linestyle='--')
plt.legend()
plt.show()
```

The vertical line at 500 calories helps identify how many items are above or below that threshold. The KDE curve gives a better understanding of the overall shape of the data distribution.

```python
plt.figure(figsize=(10,6))
sns.boxplot(df[['Total Fat','Saturated Fat']])
plt.title('Box Plot of Total Fat & Saturated Fat')
plt.xlabel('Fat Type')
plt.ylabel('Grams')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Cholesterol',y='Item',data=df.sort_values('Cholesterol',ascending=False).head(10), palette='mako')
plt.title('Top 10 items with highest cholesterol content ')
plt.xlabel('Cholesterol(mg)')
plt.ylabel('Item')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Protein',y='Item',data = df.sort_values('Protein',ascending=False).head(10),palette='rocket')
plt.title('Top 10 High Protein Items')
plt.xlabel('Protein')
plt.ylabel('Items')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data = df,kde=True,x='Sugars',bins=30)
plt.xlabel=('Sugars')
plt.ylabel=('Frequency')
plt.title('Distribution of Sugar Content')
plt.axvline(x=10,color='r',linestyle='-',label='10g Sugars')
plt.legend()
plt.show()
```

```python
# Ensure proper data types
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
df['Total Fat'] = pd.to_numeric(df['Total Fat'], errors='coerce')
df['Cholesterol'] = pd.to_numeric(df['Cholesterol'], errors='coerce')
df['Saturated Fat'] = pd.to_numeric(df['Saturated Fat'], errors='coerce')
df['Carbohydrates'] = pd.to_numeric(df['Carbohydrates'], errors='coerce')
df['Protein'] = pd.to_numeric(df['Protein'], errors='coerce')
df['Sugars'] = pd.to_numeric(df['Sugars'], errors='coerce')
```

```python
# Cholesterol
highest_cholesterol = df.loc[df['Cholesterol'].idxmax()]
print(f"Highest Cholesterol: {highest_cholesterol['Item']} with {highest_cholesterol['Cholesterol']} mg")
```

```python
highest_saturated_fat = df.nlargest(2, 'Saturated Fat')
print(f"Highest Saturated Fats: {highest_saturated_fat[['Item', 'Saturated Fat']]}")
```

```python
# Protein vs Caloric
plt.figure(figsize=(10,6))
sns.scatterplot(x='Calories',y='Protein',data=df)
plt.axvline(x=200,color='y',linestyle='--',label='2000 calories')
plt.axvline(x=500,color='b',linestyle='--',label='2500 calories')
plt.axhline(y=56,color='r',linestyle='--',label='56g protein')
plt.title('Calories vs Protein')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.legend()
plt.show()
```

```python
# Fat vs Calories
plt.figure(figsize=(8,6))
sns.scatterplot(x='Calories',y='Total Fat',data=df)
plt.axvline(x=2000,color='b',linestyle='--',label='2000 calories')
plt.axvline(x=2500,color='r',linestyle='--',label='2500 calories')
plt.axhline(y=78,color='g',linestyle='--',label='78g fat')
plt.title('Fat vs Calories')
plt.xlabel('Calories')
plt.ylabel('Total Fat')
plt.legend()
plt.show()
```

```python
# Sugar vs Calories
plt.figure(figsize=(10,6))
sns.scatterplot(x='Calories',y='Sugars',data = df)
plt.axvline(x=1000, color='r',linestyle='--',label='1000 calories')
plt.axhline(y=25, color='b',linestyle='--',label='25g sugars(Womens)')
plt.axhline(y=36, color='g',linestyle='--',label='36g sugars(Mens)')
plt.title('Sugar vs Calories')
plt.xlabel('Calories')
plt.ylabel('Sugars')
plt.legend()
plt.show()
```
