```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance=pd.read_csv('insurance.csv')
insurance.head()
```

```python
insurance.corr(numeric_only=True)
```

### Age and charges have the strongest relationship followed by BMI and charges. #for categorical variable for bar plot or pie chart.smoker, gender,region

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(10,8))
insurance.plot(kind='hist',y='age',bins=70,color='b',ax=axes[0][0])
insurance.plot(kind='hist',y='bmi',bins=70,color='y',ax=axes[0][1])
insurance.plot(kind='hist',y='children',bins=70,color='g',ax=axes[1][0])
insurance.plot(kind='hist',y='charges',bins=70,color='r',ax=axes[1][1])
plt.show()
```

### Out of all the numeric variables,only BMI is distributed normally.

```python
sns.scatterplot(x='bmi',y='charges',data=insurance,hue='smoker')
plt.show()
```

```python
sns.pairplot(insurance,hue='smoker')
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('insurance.csv')
```

```python
# Calculate the sizes for each sex
sizes_sex = df['sex'].value_counts()
print(sizes_sex)
```

```python
sizes_region = df['region'].value_counts()
print(sizes_region)
```

```python
sizes_smoker = df['smoker'].value_counts()
print(sizes_smoker)
```

```python
import matplotlib.pyplot as plt


labels_region = sizes_region.index
sizes_region = sizes_region.values
colors_region = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.figure(figsize=(8, 8))
plt.pie(sizes_region, labels=labels_region, colors=colors_region, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution by Region')
plt.show()
```

```python
# Plotting pie chart for smoker
labels_smoker = sizes_smoker.index
sizes_smoker = sizes_smoker.values
colors_smoker = ['gold', 'lightcoral']

plt.figure(figsize=(6, 6))
plt.pie(sizes_smoker, labels=labels_smoker, colors=colors_smoker, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution by Smoker Status')
plt.show()
```
